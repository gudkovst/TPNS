```python
import numpy as np
import pandas as pd
import seaborn as sns

def encoding(data: pd.DataFrame) -> pd.DataFrame:
    data_out = data.copy()
    for col in data_out:
        data_out[col] = data_out[col].apply(lambda x: ord(x) - ord('a'))
    return (data_out - data_out.min()) / (data_out.max() - data_out.min())
```

```python
data = pd.read_csv("C:\\Users\\User\\PycharmProjects\\lab2\\agaricus-lepiota.csv", header=None)
data = encoding(data)
```

```python
L = len(data.index)
CN = data.count()  # количество
NP = ((L - CN) / L) * 100  # процент пропущенных значений
MIN = data.min()  # минимум
Q1 = data.quantile(q=0.25)  # первый квартиль
MEAN = data.mean()  # среднее значение
MED = data.median()  # медиана
Q3 = data.quantile(q=0.75)  # третий квартиль
MAX = data.max()  # максимум
STD = data.std()  # стандартное отклонение
P = data.nunique()  # мощность
IQ = Q3 - Q1  # интерквартильный размах

frame = pd.concat([CN, NP, MIN, Q1, MEAN, MED, Q3, MAX, STD, P, IQ], axis=1, join="inner")
frame = frame.T
f = pd.DataFrame(frame)
f.index = ['Количество', 'Процент пропусков', 'Минимум', 'Первый квартиль', 'Среднее', 'Медиана', 'Третий квартиль',
           'Максимум', 'Стандартное отклонение', 'Мощность', 'Интерквартильный размах']
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
f
```
