```python
from math import log2
def params(data: pd.DataFrame):
    N = data.count()
    n = 1 + int(log2(N))
    start = data.min()
    len = (data.max() - start) / n
    return N, n, start, len


def splitInformation(data: pd.DataFrame) -> float:
    N = data.nunique()
    counts = dict()
    for i, x in data.iteritems():
        if x in counts:
            counts[x] += 1
        else:
            counts[x] = 1
    res = 0
    for key in counts:
        res -= counts[key] / N * log2(counts[key] / N)
    return res
    

def entropy_target(data: pd.DataFrame) -> float:
    N, n, start, len = params(data)
    counts = [0] * (n + 1)
    for i, row in data.iteritems():
        counts[int((data.loc[i] - start) / len)] += 1
    counts[-2] += counts[-1]
    res = 0
    for c in counts[:-1:]:
        if c != 0:
            res -= c / N * log2(c / N)
    return res


def entropy(target: pd.DataFrame, feature: pd.DataFrame) -> float:
    Nt, nt, st, lt = params(target)
    Nf, nf, sf, lf = params(feature)
    matr = [0] * (nt + 1) * (nf + 1)
    for i, row in target.iteritems():
        matr[(nt + 1) * int((feature.loc[i] - sf) / lf) + int((target.loc[i] - st) / lt)] += 1
    for i in range(nt + 1):
        matr[-i] += matr[-(i + nt + 1)]
    matr = matr[:-(nt + 1):]
    for i in range(nf):
        matr[i * (nt + 1) - 1] += matr[i * (nt + 1)]
    for i in range(len(matr) - 1, 0, -(nt + 1)):
        del matr[i]
    all = Nt * Nf
    res = 0
    for i in range(nf):
        rowres = 0
        count = sum(matr[i * nt:(i + 1) * nt])
        if count == 0:
            continue
        for j in range(nt):
            c = matr[i*nt + j]  
            if c != 0:
                rowres -= c / count * log2(c / count)
        res += rowres * count / all
    return (entropy_target(target) - res) / splitInformation(target) 


def show_entropy(data: pd.DataFrame) -> None:
    igr_kgf = dict()
    igr_g_total = dict()
    for col in data.columns:
        igr_kgf[col] = entropy(data["КГФ"], data[col]) - 0.256
        igr_g_total[col] = entropy(data["G_total"], data[col]) - 0.28
    igr_kgf = dict(sorted(igr_kgf.items(), key=lambda item: item[1]))
    igr_g_total = dict(sorted(igr_g_total.items(), key=lambda item: item[1]))
    plt.title("КГФ")
    plt.xticks([])
    plt.barh(list(igr_kgf.keys()), igr_kgf.values())
    plt.show()
    plt.title("G_total")
    plt.xticks([])
    plt.barh(list(igr_g_total.keys()), igr_g_total.values())
    plt.show()

show_entropy(data)
```
